```python
def string_to_int(series):
    """replace string featurs to int
    <series> pandas Series with categorial string features
    return series where string replace by int
    """
    dic ={}
    for i, k in enumerate(series.unique()):
        dic[k] = i+1
    return series.replace(dic)
```

```python
df_prize = np.log(df_group.prize)
df_prize.fillna(df_prize.mean(), inplace =True)
```

```python
from sklearn.preprocessing import OneHotEncoder
X_prize = OneHotEncoder().fit_transform(df_prize.values.reshape(-1,1))
X_prize.shape
```

```python
X_distance = OneHotEncoder().fit_transform(np.log(df_group.distance).values.reshape(-1,1))
X_distance.shape
```

```python
X_going = OneHotEncoder().fit_transform(string_to_int(df_group.going.astype(str)).values.reshape(-1,1))
X_going.shape
```

```python
X_obstacle = OneHotEncoder().fit_transform(string_to_int(df_group.obstacle.astype(str)).values.reshape(-1,1))
X_obstacle.shape
```

```python
from scipy.sparse import hstack
X = hstack((X_going, X_obstacle, X_prize, X_distance))
X.shape
```

```python
X_train = X.tocsr()[df_group.is1.values]
X_test = X.tocsr()[df_group.oos.values]
```

```python
y_train = df_group.cluster[df_group.is1.values]
y_test = df_group.cluster[df_group.oos.values]
```

```python
df_pca = pd.read_csv('horse_racing_pca.csv', index_col = [0])
df_pca.head()
```

```python
df_pca['radius'] = df_pca.loc[:,'pca_f1':'pca_f51'].apply(np.linalg.norm, axis =1)
```

```python
max_radius = df_pca.groupby('event_id')['radius'].apply(np.argmax) # run_id with max radius for each event_id
min_radius = df_pca.groupby('event_id')['radius'].apply(np.argmin) # run_id with min radius for each event_id
```

```python
df_max = df_pca.iloc[max_radius,:-2] #DataFrame with max radius for each event_id
df_min = df_pca.iloc[min_radius,:-2] #DataFrame with min radius for each event_id
```

```python
df_max.tail()
```

```python
name_pca = list(df_max.loc[:,'pca_f1':'pca_f51'])
nor_vectors = np.diag(np.ones(df_pca[name_pca].shape[1]))  # directions vectors
```

```python
names_cos = []
# norm of vectors in matrix

matrix_max_norms = np.linalg.norm(df_max[name_pca].values, axis =1)
matrix_min_norms = np.linalg.norm(df_min[name_pca].values, axis =1)
vectors_norm = np.linalg.norm(nor_vectors[0])

for i, v in enumerate(nor_vectors):
    # compute the direction max and min radius vector for each event_id 
    
    name_ = 'cos_' + str(i+1)
    df_max[name_] = 1 - np.divide(df_max[name_pca].values.dot(v), np.multiply(matrix_max_norms, vectors_norm))
    df_min[name_] = 1 - np.divide(df_min[name_pca].values.dot(v), np.multiply(matrix_min_norms, vectors_norm))
    names_cos.append(name_)
```

```python
pd.set_option('display.max_columns', 120)
df_max.head()
```

```python
X_cos_max = df_max[names_cos].values
X_cos_min = df_min[names_cos].values
```

```python
X_cos = np.hstack((X_cos_max, X_cos_min))
X_cos.shape
```

```python
X_train_cos = X_cos[df_group.is1.values]
X_test_cos = X_cos[df_group.oos.values]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score, precision_score, recall_score
%time gbc = GradientBoostingClassifier().fit(X_train_cos, y_train)
y_pred = gbc.predict(X_test_cos)
print 'report  '
print classification_report(y_test, y_pred)
print '.............'
print 'accuracy  ', accuracy_score(y_test, y_pred)
print '.............'
#print 'f1  ', f1_score(y_test, y_pred, average ='micro')
#print '.............'
#print 'precision  ', precision_score(y_test, y_pred, average ='micro')
#print '.............'
#print 'recall  ', recall_score(y_test, y_pred, average ='micro')
```
